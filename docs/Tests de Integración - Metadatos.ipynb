```python
import pandas as pd
import requests
```

```python
BASE_URL = None  # Setearla con una URL válida
METADATA_URL = 'http://infra.datos.gob.ar/catalog/modernizacion/dataset/1/distribution/1.2/download/series-tiempo-metadatos.csv'
ENDPOINT_URL = BASE_URL + 'series/api/search/'
```

```python
series_metadata = pd.read_csv(METADATA_URL)
series_metadata.tail(5)
```

```python
def api_search(**kwargs):
    return requests.get(ENDPOINT_URL, params=kwargs).json()
```

```python
api_search(q='motos')
```

## Búsqueda de "basura" para medir performance

```python
%%time

for i in range(100):
    api_search(str(i))
```

## Uso del endpoint

Agarramos una serie random conocida, y luego buscamos su descripción en la API. Esperamos que esté en los resultados

```python
random_series = series_metadata.sample(1)
random_series
```

```python
random_series['serie_descripcion']
```

```python
results = api_search(q=random_series['serie_descripcion'], limit=1000)

result_ids = set([x['field']['id'] for x in results['data']])
random_series['serie_id'].values[0] in result_ids  # True si la serie random está en los resultados
```
